```python
import csv
import matplotlib.pyplot as plt
print("Files Available for input: CA1.csv,CA2.csv,CA3.csv,TUTORIAL.csv,SEM.csv ")
def rollnum():
    print("Roll numbers available: 1,2,3,4,5")
    f=open("E:\\PYTHON\\CA1.csv",'r')
    g=open("E:\\PYTHON\\CA2.csv",'r')
    h=open("E:\\PYTHON\\CA3.csv",'r')
    i=open("E:\\PYTHON\\TUTORIAL.csv",'r')
    j=open("E:\\PYTHON\\SEM.csv",'r')
    r1=csv.reader(f)
    r2=csv.reader(g)
    r3=csv.reader(h)
    r4=csv.reader(i)
    r5=csv.reader(j)
    finaloutput= open(r'E:\PYTHON\finaloutput.csv','w',newline='')
    o1=csv.writer(finaloutput)
    rollno=input('Enter Roll Number: ')
    for it in r1:
        if rollno==it[0]:
            o1.writerow('REPORT CARD')
            o1.writerow(['Roll No',rollno])
            o1.writerow(['Tests','Maths','Science','English','SST','2Lang','Total','Percentage'])
            print('REPORT CARD')
            print(['Roll Number',rollno])
            print(['Tests','Maths','Science','English','SST','2Lang','Total','Percentage'])
            it.pop(0)
            it.pop(0)
            it.insert(0,'CA 1')
            o1.writerow(it)
            print(it)
            break
    else:
        print('Oops!Invalid roll number, Try again')
    for it in r2:
        if rollno==it[0]:
            it.pop(0)
            it.pop(0)
            it.insert(0,'CA 2')
            o1.writerow(it)
            print(it)
    for it in r3:
        if rollno==it[0]:
            it.pop(0)
            it.pop(0)
            it.insert(0,'CA 3')
            o1.writerow(it)
            print(it)
    for it in r4:
        if rollno==it[0]:
            it.pop(0)
            it.pop(0)
            it.insert(0,'TUTORIAL')
            o1.writerow(it)
            print(it)
    for it in r5:
        if rollno==it[0]:
            it.pop(0)
            it.pop(0)
            it.insert(0,'SEMESTER')
            o1.writerow(it)
            print(it)
    f.close()
    g.close()
    h.close()
    i.close()
    j.close()

def namewise():
    print("Names Available: Aakash, Akash, Dayanitha, Anirudha,Deishaun")
    f=open("E:\\PYTHON\\CA1.csv",'r')
    g=open("E:\\PYTHON\\CA2.csv",'r')
    h=open("E:\\PYTHON\\CA3.csv",'r')
    i=open("E:\\PYTHON\\TUTORIAL.csv",'r')
    j=open("E:\\PYTHON\\SEM.csv",'r')
    r1=csv.reader(f)
    r2=csv.reader(g)
    r3=csv.reader(h)
    r4=csv.reader(i)
    r5=csv.reader(j)
    finaloutput= open(r'E:\PYTHON\finaloutput.csv','w',newline='')
    o1=csv.writer(finaloutput)
    name=input('Enter Name: ')
    for it in r1:
        if name.lower()==it[1].lower():
            o1.writerow('REPORT CARD')
            o1.writerow(['Name',name.upper()])
            o1.writerow(['Tests','Maths','Science','English','SST','2Lang','Total','Percentage'])
            print('REPORT CARD')
            print(['Name',name.upper()])
            print(['Tests','Maths','Science','English','SST','2Lang','Total','Percentage'])
            it.pop(0)
            it.pop(0)
            it.insert(0,'CA 1')
            o1.writerow(it)
            print(it)
            break
    else:
        print('Oops!Invalid Name, Try again')
    for it in r2:
        if name.lower()==it[1].lower():
            it.pop(0)
            it.pop(0)
            it.insert(0,'CA 2')
            o1.writerow(it)
            print(it)
    for it in r3:
        if name.lower()==it[1].lower():
            it.pop(0)
            it.pop(0)
            it.insert(0,'CA 3')
            o1.writerow(it)
            print(it)
    for it in r4:
        if name.lower()==it[1].lower():
            it.pop(0)
            it.pop(0)
            it.insert(0,'TUTORIAL')
            o1.writerow(it)
            print(it)
    for it in r5:
        if name.lower()==it[1].lower():
            it.pop(0)
            it.pop(0)
            it.insert(0,'SEM')
            o1.writerow(it)
            print(it)
    f.close()
    g.close()
    h.close()
    i.close()
    j.close()

def subwise():
    print("Subjects available:Eng ,SST,Maths,Sci,2L")
    f = open("E:\\PYTHON\\CA1.csv", 'r')
    g = open("E:\\PYTHON\\CA2.csv", 'r')
    h = open("E:\\PYTHON\\CA3.csv", 'r')
    i = open("E:\\PYTHON\\TUTORIAL.csv", 'r')
    j = open("E:\\PYTHON\\SEM.csv", 'r')
    r1 = csv.reader(f)
    r2 = csv.reader(g)
    r3 = csv.reader(h)
    r4 = csv.reader(i)
    r5 = csv.reader(j)

    finaloutput = open(r'E:\PYTHON\finaloutput.csv', 'w', newline='')
    o1 = csv.writer(finaloutput)

    subject = input('Enter Subject: ')
    d = {'maths': 2, 'sci': 3, 'eng': 4, 'sst': 5, '2l': 6}

    if subject.lower() not in d:
        print('Oops! Invalid subject, Try again')
        return

    o1.writerow(['Test', 'CA 1', 'CA 2', 'CA 3', 'TUTORIAL', 'SEM'])

    sub_index = d[subject.lower()]

    for row in zip(r1, r2, r3, r4, r5):
        roll_no = row[0][0]
        marks = [roll_no] + [r[sub_index] for r in row]
        o1.writerow(marks)
    f.close()
    g.close()
    h.close()
    i.close()
    j.close()

def plot_marks():
    print("Files Available for input: CA1.csv,CA2.csv,CA3.csv,TUTORIAL.csv,SEM.csv ")
    print("Subjects available:Eng ,SST,Maths,Sci,2L")
    subject = input('Enter Subject to Plot: ').lower()
    file_name = input('Enter File Name to Plot: ').strip().lower()

    if subject not in ['maths', 'sci', 'eng', 'sst', '2l']:
        print('Invalid subject!')
        return
    if file_name not in ['ca1.csv', 'ca2.csv', 'ca3.csv', 'tutorial.csv', 'sem.csv',]:
        print('Invalid file name!')
        return

    with open(f"E:\\PYTHON\\{file_name}", 'r') as file:
        reader = csv.reader(file)
        next(reader)
        data = {row[0]: int(row[2]) for row in reader}

    plt.figure(figsize=(10, 6))
    plt.bar(data.keys(), data.values(), color='skyblue')
    plt.xlabel('Roll Number')
    plt.ylabel('Marks')
    plt.title(f'Marks in {subject.upper()}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def inputs():
    expected_files = ['CA1.csv', 'CA2.csv', 'CA3.csv', 'TUTORIAL.csv', 'SEM.csv']
    global dupfile
    dupfile = []
    for _ in range(5):
        fileinput = input('Enter File Name(.csv)= ').strip().lower()
        if fileinput.lower() not in [f.lower() for f in expected_files]:
            print(f'Invalid file name: {fileinput}. Please enter one of the following: {", ".join(expected_files)}')
            return False
        dupfile.append(fileinput)
    return True

finaloutput = open(r'E:\PYTHON\finaloutput.csv', 'w', newline='')
o1 = csv.writer(finaloutput)

if inputs():
    while True:
        ch = input('''
        Enter search method
        1- Roll number
        2- Name
        3- Subwise
        4- Plot marks
        5- Exit''')
        if ch == '1':
            rollnum()
        elif ch == '2':
            namewise()
        elif ch == '3':
            subwise()
        elif ch == '4':
            plot_marks()
        elif ch == '5':
            break
        else:
            print('Oops!Invalid option, Try again')
else:
    print('Invalid file name(s), please try again with correct file name(s)')
```
